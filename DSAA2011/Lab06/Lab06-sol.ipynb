**Question 1.** Given a sample $\mathbf{x}$ and the parameters $(\boldsymbol{\theta}, \theta_0)$, assume the label $y\in\{-1, 1\}$ and 
$$
\log \frac{\operatorname{Pr}\left(y=1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0\right)}{\operatorname{Pr}\left(y=-1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0\right)}=\langle\boldsymbol{\theta}, \mathbf{x}\rangle+\theta_0.
$$
Show that
$$
    \Pr \left(y=1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0\right)=g\left(\langle\boldsymbol{\theta}, \mathbf{x}\rangle+\theta_0\right)=\frac{1}{1+\exp \left(-\left(\langle\boldsymbol{\theta}, \mathbf{x}\rangle+\theta_0\right)\right)}. 
$$

**Solution**

---

<!-- #### **1. Hand Calculation of** -->


Since probabilities add to 1:
$$
\operatorname{Pr}(y = -1 \mid  \mathbf{x}, \boldsymbol{\theta}, \theta_0 ) = 1 - \operatorname{Pr}(y = 1 \mid  \mathbf{x}, \boldsymbol{\theta}, \theta_0 )
$$ 

We have:

$$
\log \frac{\operatorname{Pr}(y =1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0)}{ 1 - \operatorname{Pr}(y = 1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0)} = \langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0
$$

Then:

$$
\frac{\operatorname{Pr}(y =1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0)}{ 1 - \operatorname{Pr}(y = 1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0)} =  \mathrm{exp}(\langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0)
$$

Hence:

$$
1 + \frac{\operatorname{Pr}(y =1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0)}{ 1 - \operatorname{Pr}(y = 1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0)} = 1 + \mathrm{exp}(\langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0)
$$

Then:

$$
\frac{1}{ 1 - \operatorname{Pr}(y = 1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0)} = 1 + \mathrm{exp}(\langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0)
$$

Then:

$$
1 - \operatorname{Pr}(y = 1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0) = \frac{1}{1 + \mathrm{exp}(\langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0)}
$$

Then:

$$
\operatorname{Pr}(y = 1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0) = 1 - \frac{1}{1 + \mathrm{exp}(\langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0)}
$$

Then:

$$
\operatorname{Pr}(y = 1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0) = \frac{\mathrm{exp}(\langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0)}{1 + \mathrm{exp}(\langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0)}
$$

Finally, by multiplying $\mathrm{exp}(-(\langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0))$ for both the numerator and denominator of the fraction on the right side of the euqtion, we have:

$$
\operatorname{Pr}(y = 1 \mid \mathbf{x}, \boldsymbol{\theta}, \theta_0) = \frac{1}{1 + \mathrm{exp}(-(\langle \boldsymbol{\theta},\mathbf{x} \rangle + \theta_0))}
$$

**Question 2.** To find the MLE of parameters for logistic regression, prove the gradients with respect to $\boldsymbol{\theta}$ and $\theta_0$ are as follows:
\begin{align}
\frac{\mathrm{d}}{\mathrm{d}\theta_0}\log \big[1+\exp(-y_t( \langle\mathbf{x}_t,  \boldsymbol{\theta}\rangle+\theta_0 ) )\big]&=  -y_t\frac{\exp \big(-y_t( \langle\mathbf{x}_t,  \boldsymbol{\theta}\rangle+\theta_0 )  \big)}{1+\exp \big(-y_t( \langle\mathbf{x}_t,  \boldsymbol{\theta}\rangle+\theta_0 )  \big)}  \\
&=-y_t \big[ 1-\Pr(y_t\mid \mathbf{x}_t, \boldsymbol{\theta}, \theta_0 ) \big],\\
\frac{\mathrm{d}}{\mathrm{d}\boldsymbol{\theta}}\log \big[1+\exp(-y_t( \langle\mathbf{x}_t,  \boldsymbol{\theta}\rangle+\theta_0 ) )\big]&=-y_t \mathbf{x}_t\big[ 1-\Pr(y_t\mid \mathbf{x}_t, \boldsymbol{\theta}, \theta_0 ) \big].
\end{align}

**Solution**

---
$$
\begin{align}
\frac{\mathrm{d}}{\mathrm{d}\theta_0} \log \big[ 1 + \exp( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) \big]
    &= \frac{1}{ 1 + \exp( -y_t ( \langle\mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) } \frac{\mathrm{d}}{\mathrm{d}\theta_0} \exp( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) \\
    &= \frac{ \exp( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) }{ 1 + \exp( -y_t ( \langle\mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) } \frac{\mathrm{d}}{\mathrm{d}\theta_0} ( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) \\
    &= -y_t \frac{ \exp( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) }{ 1 + \exp( -y_t ( \langle\mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) } \\
    &= -y_t \frac{ \exp( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) + 1 - 1 }{ 1 + \exp( -y_t ( \langle\mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) } \\
    &= -y_t \big[ 1 - \frac{1}{ 1 + \exp( -y_t ( \langle\mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) } \big] \\
    &= -y_t \big[ 1- \Pr( y_t \mid \mathbf{x}_t, \boldsymbol{\theta}, \theta_0 ) \big]
\end{align}
$$

Similarly, we have:

$$
\begin{align}
\frac{\mathrm{d}}{\mathrm{d}\boldsymbol{\theta}} \log \big[ 1 + \exp( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) \big]
    &= \frac{1}{ 1 + \exp( -y_t ( \langle\mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) } \frac{\mathrm{d}}{\mathrm{d}\boldsymbol{\theta}} \exp( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) \\
    &= \frac{ \exp( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) }{ 1 + \exp( -y_t  ( \langle\mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) } \frac{\mathrm{d}}{\mathrm{d}\boldsymbol{\theta}} ( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) \\
    &= -y_t \mathbf{x}_t \frac{ \exp( -y_t ( \langle \mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) }{ 1 + \exp( -y_t ( \langle\mathbf{x}_t, \boldsymbol{\theta} \rangle + \theta_0 ) ) } \\
    &= -y_t \mathbf{x}_t \big[ 1- \Pr( y_t \mid \mathbf{x}_t, \boldsymbol{\theta}, \theta_0 ) \big]
\end{align}
$$

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1. Load the dataset

```python
# sns.set_theme(style="whitegrid")
sns.set_theme()

# Load the dataset
data_path = 'Drug.csv'
data = pd.read_csv(data_path)

# Display the first few entries of the dataset
print(data.head())
```

#### 2. Explore the dataset
See the feature distribution in the dataset.

```python
# Distribution of Categorical Features

fig, ax = plt.subplots(2, 2, figsize=(8, 6))
sns.countplot(x='Sex', data=data, ax=ax[0, 0])
ax[0, 0].set_title('Sex Distribution')

sns.countplot(x='BP', data=data, ax=ax[0, 1])
ax[0, 1].set_title('Blood Pressure (BP) Distribution')

sns.countplot(x='Cholesterol', data=data, ax=ax[1, 0])
ax[1, 0].set_title('Cholesterol Distribution')

sns.countplot(x='Drug', data=data, ax=ax[1, 1])
ax[1, 1].set_title('Drug Type Distribution')

plt.tight_layout()
```

```python
# Distribution of Numerical Features, for Different Drug Types

fig, ax = plt.subplots(3, 1, figsize=(4, 9))

# Boxplot for Age vs Drug
sns.boxplot(x='Drug', y='Age', data=data, ax=ax[0])
ax[0].set_title('Age Distribution for Each Drug Type')

# Boxplot for Na vs Drug
sns.boxplot(x='Drug', y='Na', data=data, ax=ax[1])
ax[1].set_title('Sodium (Na) Level Distribution for Each Drug Type')

# Boxplot for K vs Drug
sns.boxplot(x='Drug', y='K', data=data, ax=ax[2])
ax[2].set_title('Potassium (K) Level Distribution for Each Drug Type')

plt.tight_layout()
```

#### 3. Data Preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Defining the feature set and the target variable
X = data.drop('Drug', axis=1)
y = data['Drug']

# Identifying numerical and categorical features
numerical_features = ['Age', 'Na', 'K']
categorical_features = ['Sex', 'BP', 'Cholesterol']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2025)

# Defining a column transformer with one-hot encoding for categorical features and scaling for numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

# Fitting the transformer on the training data and transforming both the training and test data
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

#### 4. Train and Test the Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize the models
lr_classifier = LogisticRegression(max_iter=1000)  # Increased max_iter for convergence

# Train the model
lr_classifier.fit(X_train_processed, y_train)

# Make predictions on the test set
y_pred = lr_classifier.predict(X_test_processed)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)

accuracy
```

#### 5. Detailed Analysis of the Model Performance

```python
from sklearn.metrics import confusion_matrix, classification_report
from tabulate import tabulate
# Initializing a dictionary to store evaluation metrics
model_evaluation_metrics = {}

# Predict on the test set
y_pred = lr_classifier.predict(X_test_processed)

# Calculate confusion matrix and classification report
conf_matrix = confusion_matrix(y_test, y_pred)
classif_report = classification_report(y_test, y_pred)

# Storing the results
model_evaluation_metrics = {
    'Confusion Matrix': conf_matrix,
    'Classification Report': classif_report
}

# Outputting the results for review
print(f"Logistic Regression Evaluation Metrics:")
print("Confusion Matrix:")
print(tabulate(conf_matrix, headers=lr_classifier.classes_, tablefmt='grid'))
print("\nClassification Report:")
print(classif_report)
```
