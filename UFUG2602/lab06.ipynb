Let ***A[1:n]*** be an array of n distinct numbers. If *i < j* and *A[i] > A[j]*, then the pair `(i, j)` is called an ***inversion*** of *A*.


1. What array with elements from the set `{1, 2, ... , n} has the most inversions? How many does it have?


2. What is the relationship between the running time of **insertion sort** and **the number of inversions** in the input array?


3. Give an algorithm that determines the number of inversions in any permutation on *n* elements in $\theta(nlgn)$ worst-case time. (Hint: Modify merge sort, if you don't use this method, please also give your method and corresponding simple explainations.)

Answers for Q1 & Q2:


1. The array ${n, n-1, n-2 ... 2, 1}$ has the most inversions: $\frac{n(n-1)}{2}$ (5% Marks)


2. The running time of insertion sort is: $\text{[The running time]} = \Theta(n + \text{[The number of inversions]})$

Answers for Q3:

```python
# Write down your method (function/class/...) here
def calcInversions(arr : list, l : int = 0, r : int = -1) -> int :
    if r == -1: 
        l = 0; r = len(arr) - 1
    if l == r: return 0
    mid = (l + r) >> 1
    p1, p2, ans = l, mid + 1, 0
    ansl, ansr = calcInversions(arr, l, mid), calcInversions(arr, mid + 1, r)
    tmp = []
    while p1 <= mid and p2 <= r:
        if arr[p1] <= arr[p2]:
            tmp.append(arr[p1]); p1 += 1
        else:
            ans += mid - p1 + 1
            tmp.append(arr[p2]); p2 += 1
    while p1 <= mid: tmp.append(arr[p1]); p1 += 1
    while p2 <= r: tmp.append(arr[p2]); p2 += 1
    for pos in range(l, r + 1) : arr[pos] = tmp[pos - l]
    return ans + ansl + ansr

# Testing usages, you should add your own count function and remeber to test all the given array case in your code answer
if __name__ == "__main__":
    # Test case 1: An array with inversions
    arr = [2, 3, 8, 6, 1]       # Should return 5
    print(calcInversions(arr))

    # Test case 2: A sorted array with no inversions
    arr = [1, 2, 3, 4, 5]       # Should return 0
    print(calcInversions(arr))

    # Test case 3: A reverse sorted array with maximum inversions
    arr = [5, 4, 3, 2, 1]       # Should return 10
    print(calcInversions(arr))
```

```python
def print_all_inversions(A):
    inversions = []
    for i in range(len(A)):
        for j in range(i + 1, len(A)):
            if A[i] > A[j]:
                inversions.append((i + 1, j + 1))  # Using 1-based indexing
    for inversion in inversions:
        print(inversion)

# Testing usages
if __name__ == "__main__":
    # Test case: An array with inversions
    arr = [2, 3, 8, 6, 1]
    print("Original array:", arr)
    print("All inversions (using 1-based indexing):")
    print_all_inversions(arr)
```
