```python
import heapq
class Graph:
	def __init__(self):
		self.graph = {} #adj list
	
	def add_edge(self, node, neighbor, weight): #add links
		if node in self.graph:
			self.graph[node].append((neighbor, weight))
		else:
			self.graph[node] = [(neighbor, weight)]

		if neighbor in self.graph:
			return
		else:
			self.graph[neighbor] = []
	
	def Dijkstra(self, start):
		'''
		Question 1: Please implement the Dijkstra algorithm to solve the shortest path problem. You should return the shorest path from the given source node to the other nodes. 
		'''
		## TODO
		dis = {}
		vis = set()
		que = [(0, start)]
		dis[start] = 0
		while len(que) > 0 :
			u = heapq.heappop(que)[1]
			if u in vis : continue
			vis.add(u)
			if u not in self.graph : continue
			for (v, w) in self.graph[u] :
				d = dis[u] + w
				if (v not in dis) or (dis[v] > d) :
					dis[v] = d
					heapq.heappush(que, (dis[v], v))
		return dis
	def BellmanFord(self, start):
		'''
		Question 2: Please implement the Bellman Ford algorithm to solve the shortest path problem. You should return the shorest path from the given source node to the other nodes. If the graph has negative loop, please output 'None!'. 
		'''
		## TODO
		dis = {}
		dis[start] = 0
		ndSet = set()
		for u in self.graph.keys() :
			ndSet.add(u)
			for (v, w) in self.graph[u] :
				ndSet.add(v)
		n = len(ndSet)
		for i in range(n) :
			for u in self.graph.keys() :
				if u not in dis : continue
				for (v, w) in self.graph[u] :
					d = dis[u] + w
					if (v not in dis) or dis[v] > d :
						dis[v] = d
		for u in self.graph.keys() :
			if u not in dis : continue
			for (v, w) in self.graph[u] :
				d = dis[u] + w
				if v in dis and dis[v] > d :
					return None
		return dis



	

graph = Graph()
graph.add_edge('1', '2', 2) #test graph. You are free to modify
graph.add_edge('2', '3', 2)
graph.add_edge('2', '4', 1)
graph.add_edge('1', '3', 5)
graph.add_edge('3', '4', 3)
graph.add_edge('1', '4', 4)
print('====Dijkstra====')
print(graph.Dijkstra('1'))
graph2 = Graph()
graph2.add_edge('1', '2', -1)
graph2.add_edge('2', '3', -1)
graph2.add_edge('3', '1', 5)
graph2.add_edge('1', '4', 3)
graph2.add_edge('3', '4', 3)
graph2.add_edge('3', '5', 3)
print('====BellmanFord====')
print(graph2.BellmanFord('1'))
graph2.add_edge('5', '1', -2)
print(graph2.BellmanFord('1'))



```
